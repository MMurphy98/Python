## 1.1 基本用法

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-3, 3, 50)
y1 = 2*x + 1
y2 = x**2

plt.figure(num=1)
plt.plot(x, y1)
plt.show()

plt.figure(num=2,figsize=(8,5))
plt.plot(x, y2)
plt.plot(x, y1, color='red', linewidth=2.0, linestyle='--')
plt.show()
```

```python
plt.figure()
plt.plot(x, y2)
plt.plot(x, y1, color='red', linewidth=2.0, linestyle='--')
plt.xlim((-1, 2))
plt.ylim((-2, 3))

plt.xlabel("Open-loop Gain")
plt.ylabel("SINAD")

new_ticks= np.linspace(-1, 2, 5)
print(new_ticks)
plt.xticks(new_ticks)
plt.yticks([-1.5, -1, 0, 1, 2.8,],
           ['really bad', 'bad', 'normal', 'good', 'really good', ])

plt.show()
```
